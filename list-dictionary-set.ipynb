```python
#creating a list
mylist=[4,0,1,3,6,5]
#adding an element to list
mylist.append(9)
#removing an element from list
mylist.remove(4)
#modifying an element in list
mylist[5]=3
print("updated list:",mylist)
```

```python
#creating a dictionary
mydict={'movie':'pushpa2','rating':4,'review':'good'}
#adding
mydict['hero']='allu arjun'
#removing 
del mydict['rating']
#modifying
mydict['review']='blockbuster'
print("upadted dictionary:",mydict)

```

```python
#creating a set
myset={4,4,0,1}
#adding
myset.add(2)
#removing
myset.remove(4)
#modifying
myset.add(5)
print("updated set:",myset)
```
